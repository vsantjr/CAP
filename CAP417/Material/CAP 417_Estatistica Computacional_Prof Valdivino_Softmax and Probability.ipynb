## Softmax in Python
----

Firstly, softmax in Python.


```python
import numpy as np
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

logits = [3.2, 5.1, -1.7]
print('Softmax - Python: ', softmax(logits))
```

## Softmax in Pytorch
----

Pytorch is a deep learning (DL) framework. Softmax is a non-linear activation function.

```python
import torch
import torch.nn as nn
```

```python
a = nn.Softmax(dim=0)
input = torch.FloatTensor(logits)
output = a(input)
print('Softmax - Pytorch: ', output)
```

## LogSoftmax in Pytorch
----

LogSoftmax is one loss function. 

```python
ls = nn.LogSoftmax()
output = -ls(input)
print('LogSoftmax:' , output)
```
